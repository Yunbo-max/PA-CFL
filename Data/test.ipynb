```python
import pandas as pd

# Create a sample DataFrame
data = {
    'SpanishColumn1': ['Hola', 'Mundo', 'Cómo', 'Estás', 'Adiós'],
    'SpanishColumn2': ['Buenos días', 'Buenas tardes', 'Buenas noches', 'Hasta luego', 'Gracias']
}

df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
file_path = 'sample_excel.xlsx'
df.to_excel(file_path, index=False)

print(f"Sample Excel file created at {file_path}")
```

```python
import pandas as pd
from googletrans import Translator

# Initialize the translator
translator = Translator()

# Read the Excel file
file_path = 'sample_excel.xlsx'
df = pd.read_excel(file_path)

# Specify the columns you want to translate
column1 = 'SpanishColumn1'
column2 = 'SpanishColumn2'

# Function to translate text
def translate_text(text):
    try:
        translation = translator.translate(text, src='es', dest='en')
        return translation.text
    except Exception as e:
        print(f"Error translating text: {text}, Error: {e}")
        return text

# Apply the translation function to the specified columns
df['EnglishColumn1'] = df[column1].apply(translate_text)
df['EnglishColumn2'] = df[column2].apply(translate_text)

# Save the translated data back to an Excel file
output_file_path = 'translated_sample_excel.xlsx'
df.to_excel(output_file_path, index=False)

print(f"Translation completed and saved to {output_file_path}")
```

```python
import pandas as pd
from googletrans import Translator

# Initialize the translator
translator = Translator()

# Read the CSV file with the specified encoding
file_path = 'DataCoSupplyChainDataset.csv'
df = pd.read_csv(file_path, encoding='latin1')

# Specify the columns you want to translate
columns_to_translate = ['Order City', 'Order Country', 'Order State']

# Function to translate text to English
def translate_text(text):
    try:
        if pd.isna(text):  # Check if the cell is NaN
            return text
        translation = translator.translate(text, dest='en')
        return translation.text
    except Exception as e:
        print(f"Error translating text: {text}, Error: {e}")
        return text

# Apply the translation function to the specified columns
for column in columns_to_translate:
    new_column_name = f"English {column}"
    df[new_column_name] = df[column].apply(translate_text)

# Save the translated data back to a CSV file
output_file_path = 'translated_supply_chain_dataset.csv'
df.to_csv(output_file_path, index=False)

print("Translation completed and saved to", output_file_path)
```
